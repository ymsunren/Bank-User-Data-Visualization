```python
import sqlite3
from pathlib import Path

Path("bank.db").touch()

conn = sqlite3.connect("bank.db")
c = conn.cursor()

c.execute(
    """CREATE TABLE bank (
      age int, job text, marital text, 
      education text, default_e text, 
      balance int, housing text, 
      loan text, contact text, 
      day int, month text, 
      duration int, campaign int, 
      pdays text, previous int, 
      poutcome text, deposit text
    ) """
)

```

```python
import pandas as pd

bank = pd.read_csv("bank.csv")
bank.rename(columns={"default": "default_e"}, inplace=True)



bank.to_sql("bank", conn, if_exists="append", index =False)
```
